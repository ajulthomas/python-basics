```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
country_df = pd.read_csv('./gapminder.csv')
```

```python
plt.scatter(country_df['gdp_cap'], country_df['life_exp'])
plt.xlabel('Gross Domestic Produce per capita')
plt.ylabel("Life Expectancy in years")
plt.title("World Development in 2007")
plt.xscale('log')
plt.show()
```
