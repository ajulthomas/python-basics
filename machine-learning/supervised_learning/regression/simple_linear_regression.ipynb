```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
# from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
diabetics_df = pd.read_csv('./datasets/diabetes_clean.csv')
diabetics_df.head()
```

```python
# simple linear regression - make predictions based on one feature

# feature variable
X = diabetics_df['bmi'].values

# target variable
y = diabetics_df['glucose'].values

print(f"Shape of y  {y.shape} {type(X)}")
print(f"Shape of X  {X.shape}")

# features must be formatted as a 2D array for scikit learn, reshaping X

X = X.reshape(-1,1)


print(f"Shape of X  {X.shape}")
```

```python
# simple linear regression
X_train,  X_test, y_train, y_test = train_test_split(X, y, train_size=0.3, random_state=42)

reg = LinearRegression()
reg.fit(X_train, y_train)
y_predict = reg.predict(X_test)
y_train_predict = reg.predict(X_train)
```

```python
plt.title("BMI vs Blood Glucose Levels")
plt.scatter(X,y,c=diabetics_df['diabetes'])
plt.plot(X_test,y_predict)
# plt.plot(X_train,y_train_predict)
plt.xlabel("Body Mass Index")
plt.ylabel("Blood Glucose (mg/dl)")
plt.show()
```

```python
reg.score(X_test,y_test)
```
